```
!pip install gym
!apt install make libopenmpi-dev zlib1g-dev
!pip install stable-baselines==2.2.0
!apt-get install python-opengl
!apt-get install -y xvfb python-opengl > /dev/null 2>&1
!pip install gym pyvirtualdisplay > /dev/null 2>&1

# Ref https://stable-baselines.readthedocs.io/en/master/modules/dqn.html

try:
  from google.colab import files
except:
  print("Could not import Google Colab.")
import matplotlib
import time
#matplotlib.use("Agg")
import numpy as np
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay
from matplotlib.image import imread
from matplotlib import rc, animation
from IPython import display
from IPython.display import HTML
%matplotlib inline

from pyvirtualdisplay import Display
display = Display(visible=0, size=(400, 300))
display.start()

import gym
from stable_baselines.common.atari_wrappers import make_atari
from stable_baselines.deepq.policies import MlpPolicy, CnnPolicy
from stable_baselines import DQN


env = make_atari('BreakoutNoFrameskip-v4')  #  Environment Load

model = DQN(CnnPolicy, env, verbose=1,learning_rate=0.05,gamma=0.99,batch_size=64) #DQN Model, policy setup with Hyper-Parameters Gamma, Epsilon, Gamma, Batch_Size
model.learn(total_timesteps=100000) # Model Learn
frames = []
i = 20000
obs = env.reset()
while i > 0:
    action, _states = model.predict(obs)  #Model Predict
    obs, rewards, done, info = env.step(action)
    fig = env.render(mode='rgb_array') #Render Environment at Current State
    frames.append(fig)
    i = i - 1

!apt install ffmpeg
!which ffmpeg
plt.rcParams['animation.ffmpeg_path'] = u'/usr/bin/ffmpeg'

fig, ax = plt.subplots() # Plot Graph
plt.axis('off')
l = ax.imshow(frames[0])

def animate(i):
    l.set_data(frames[i]) # For Animation

Writer = animation.writers['ffmpeg']
writer = Writer(fps=12, metadata=dict(artist='Me'))
ani = animation.FuncAnimation(fig, animate, frames=len(frames))

ani.save('animation.mp4', writer=writer, dpi=220) # File Safe
time.sleep(50) # let it process (only necessary in Colab)

files.download('animation.mp4')
```
